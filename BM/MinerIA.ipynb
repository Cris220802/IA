```python
import pandas as pd
import numpy as np
filename = 'DataAu.csv'
columns = ['fecha', 'turno', 'toneladas', 'cabeza_flotacion', 'concentrado_pb', 'colas_pb', 'concentrado_zn', 'colas_zn', 'concentrado_fe', 'colas_finales']
dataFrame = pd.read_csv(filename, names = columns)
dataFrame
```

## Estadística Descriptiva

```python
dataFrame.shape
```

```python
dataFrame.dtypes
```

```python
dataFrame.describe()
```

```python
dataFrameNums = dataFrame.iloc[:, 1:]
```

```python
correlation = dataFrameNums.corr(method='pearson')
correlation
```

```python
dataFrameNums.skew()
```

## Visualización de Datos

```python
import matplotlib.pyplot as plt
import seaborn as sns
dataFrameNums.iloc[:, 1:].hist(figsize=(12, 8), bins=30, grid=False)
plt.show()
```

```python
for col in dataFrame.iloc[:, 1:]:  # Excluye la primera columna si es categórica
    plt.figure(figsize=(8, 5))  # Definir tamaño de cada gráfico
    sns.kdeplot(dataFrame[col], fill=True, color="blue", linewidth=2)
    plt.title(f'Diagrama de Densidad de {col}')
    plt.xlabel("Valor")
    plt.ylabel("Densidad")
    plt.grid(True)
    plt.show()
```

```python
for col in dataFrameNums.columns:
    plt.figure(figsize=(6, 4))  # Tamaño del gráfico
    sns.boxplot(y=dataFrame[col])
    plt.title(f'Boxplot de {col}')
    plt.ylabel(col)
    plt.show()
```

```python
plt.figure(figsize = (10,10))
ax = sns.heatmap(correlation, vmax=1, square=True, annot=True, cmap='viridis')
plt.title("Matriz de correlacion")
plt.show()
```
